$$
\begin{cases}
\frac{d y}{d t} = - 2 y\\
y(0) = 1
\end{cases}
$$
solve:
$$
y(t) = e^{-2t}
$$

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Define the differential equation
def dydt(t, y):
    return - 1 * y

# Initial condition
y0 = 1

# Time span
t_span = (0, 5)
t_eval = np.linspace(0, 5, 10000)
print(t_span)

# Solve the IVP
sol = solve_ivp(dydt, t_span, [y0], t_eval=t_eval)
print(len(sol.t), len(sol.y[0]))
print('t', sol.t)
print('y', sol.y[0])

# Plot the solution
plt.plot(sol.t, sol.y[0], marker='o')
plt.xlabel('t')
plt.ylabel('y')
plt.title('Solution to the IVP')
plt.show()
```
